## Merging Data Frames

```python
#Drop unnamed columns in df_prods
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
#Drop unnamed columns in df_orders_products
df_orders_products = df_orders_products.drop(columns = ['Unnamed: 0.1', 'Unnamed: 0', '_merge'])
```

```python
#Merge df_prods and df_orders_products dataframes
df_merged = df_prods.merge(df_orders_products, on = 'product_id', indicator = True)
```
