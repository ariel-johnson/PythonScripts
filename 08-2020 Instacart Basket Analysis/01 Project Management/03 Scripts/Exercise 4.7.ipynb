## Adding Price Label

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

## If-Statements with For-Loops

```python
#Finding busiest and slowest days of the week
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

Day 0 is Saturday. Saturday is the busiest day of the week, with the most amount of orders on that day. Wednesday (4) is the slowest day of the week, with the lowest amount of orders on that day.

```python
#Creating a new column portraying the busiest day, least busy, and regularly busy
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
#Combining results with the ords_prods_merge dataframe
df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

## Task 4.7 #2. New Column containing "Busiest_Days"

```python
#Create a new column changing labels in "Busiest Day" column to "Busiest Days". With two busiest days and two slowest days.
results = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value ==1:
    results.append("Busiest days")
  elif value == 4 or value==3:
    results.append("Least busy days")
  else:
    results.append("Regularly busy")
```

```python
results
```

```python
#Combining results with the ords_prods_merge dataframe
df_ords_prods_merge['busiest_days'] = results
```

## #4. New Column Containing "Busiest_period_of_day"

```python
#Finding busiest hours of the day
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
result_2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_2.append("Fewest orders")
  else:
    result_2.append("Average orders")
```

```python
result_2
```

```python
#Incorporating new column into dataframe
df_ords_prods_merge['busiest_period_of_day'] = result_2
```
